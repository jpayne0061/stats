# Analysis of Salaries

The Notebook has added support for `raw_input` and `%debug`, as of 1.0.

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
%pylab inline

master = pd.read_csv("../data_sets/baseballdatabank-master/core/Master.csv")
salaries = pd.read_csv("../data_sets/baseballdatabank-master/core/Salaries.csv")

salary_merge = pd.merge(master, salaries, on='playerID', how='inner')
salary_height_15 = salary_merge[salary_merge['yearID'] == 2015]

salary = salary_height_15['weight']
height = (salary_height_15['salary'])/1000000

print stats.pearsonr(salary, height)

plt.scatter(height, salary)
plt.tick_params(labelsize=14)
plt.xlabel('Salary(Millions)', fontsize=18)
plt.ylabel('Weight(lbs)', fontsize=18)
plt.axis([0, 35, 150, 320])
plt.title('MLB Salary vs Weight', fontsize=22, y=1.025)
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
fig.savefig('weight_salary.png', dpi=100)
```

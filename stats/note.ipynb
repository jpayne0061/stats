```python
deck = []

for i in range(4):
    deck.append(1)
for i in range(16):
    deck.append(10)
for i in range(2, 10):
    for num in range(4):
        deck.append(i)

#first attempt without numpy or using tools for sampling
samples = []
for i in range(1000):
    samp_mean = []
    loop_deck = list(deck)
    #print str(loop_deck) + 'loop deck'
    #print str(deck) + "deck"
    for i in range(3):
        samp_mean.append(loop_deck.pop(randint(0, len(loop_deck)-1)))
    samples.append(np.mean(samp_mean))
    
plt.hist(samples, normed=True)

m = stats.mode(samples)
print "mode= " + str(m[0])
print "mean= " + str(np.mean(samples))
print "median= " + str(np.median(samples))
print "SD= " + str(np.std(samples))
print "variance= " + str(np.var(samples))
print "sample size= " + str(len(samples))

outside_one_sd = []
for i in samples:
    if (i < 4.88) or (i > 8.45):
        outside_one_sd.append(i)
print len(outside_one_sd)
print "percentage outside 1 SD, expecting 32 percent= " + str(float(len(outside_one_sd))/len(samples)*100)
```
